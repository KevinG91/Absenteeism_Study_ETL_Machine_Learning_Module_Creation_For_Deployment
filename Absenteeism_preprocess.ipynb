##### BMI Categories:
* Underweight = <18.5
* Normal weight = 18.5–24.9
* Overweight = 25–29.9
* Obesity = BMI of 30 or greater

we already know the dataframe has 700 rows so i'm gonna use that from now on

```python
underweight = []
normal_weight = []
overweight = []
obese = []

for i in range(len(df_reasons['Body Mass Index'])):
    if df_reasons['Body Mass Index'][i] < 18.5:
        underweight.append(1)
    else:
        underweight.append(0)

for i in range(len(df_reasons['Body Mass Index'])):
    if df_reasons['Body Mass Index'][i] >= 18.5 and df_reasons['Body Mass Index'][i] <= 24.9:
        normal_weight.append(1)
    else:
        normal_weight.append(0)

for i in range(len(df_reasons['Body Mass Index'])):
    if df_reasons['Body Mass Index'][i] > 24.9 and df_reasons['Body Mass Index'][i] <= 29.9:
        overweight.append(1)
    else:
        overweight.append(0)

for i in range(len(df_reasons['Body Mass Index'])):
    if df_reasons['Body Mass Index'][i] > 29.9:
        obese.append(1)
    else:
        obese.append(0)
```

```python
print(set(underweight)), print(len(underweight))
print(set(normal_weight)), print(len(normal_weight))
print(set(overweight)), print(len(overweight))
print(set(obese)), print(len(obese))
```

```python
absenteeism_3sigma = absenteeism_mean+(absenteeism_std*3)
absenteeism_3sigma
```

```python
absenteeism_3sigma = absenteeism_3sigma+absenteeism_std
absenteeism_3sigma
```

i'm not going to do anything with these, though, nothing really stands out and the absent hours aren't particularly ridiculous, i just wanted to see if i had crazy outliers

according to google it is considered excess absenteeism when the person is absent for more that 3 hours every 90 days or 12 hours in a year, the dataset shows registers from 2015 to 2018 so 3 years. Anyone with more than 36 hours absent is gonna be considered an excessive absentee

first let's get the IDs

```python
ID_list = list(raw_data.ID.unique())
```

then we initialize the dictionary that will save the hours absent

```python
absent_hours = {i:0 for i in ID_list}
```

add the hours to each employee

```python
for i in range(700):
    absent_hours[raw_data.iloc[i]['ID']] = absent_hours[raw_data.iloc[i]['ID']]+raw_data.iloc[i]['Absenteeism Time in Hours']
```

looks good

```python
absent_hours
```

now we label the excessive absentees as 1 and the rest as 0

```python
for i, (key, value) in enumerate(absent_hours.items()):
    if value > 36:
        absent_hours[key] = 1
    else:
        absent_hours[key] = 0
    
```

looks good

```python
absent_hours
```

now we can create the excessive absenteeism column

```python
excessive_absenteeism = []
for i in range(700):
    if absent_hours[raw_data.iloc[i]['ID']] == 1:
        excessive_absenteeism.append(1)
    else:
        excessive_absenteeism.append(0)
```

looks good
